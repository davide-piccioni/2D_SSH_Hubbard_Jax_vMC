```python
import sys
import numpy as np

def process_data(file_path, name):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    data = {}
    current_key = None
    for line in lines:
        line = line.strip()
        if not line or line.startswith('---'):
            continue
        if '_' in line:
            current_key = line
            data[current_key] = {'mean': [], 'err': []}
        else:
            if '±' in line:
                mean, err = line.split(' ± ')
                data[current_key]['mean'].append(float(mean))
                data[current_key]['err'].append(float(err))
            else:
                data[current_key]['mean'].append(float(line))
                data[current_key]['err'].append(0.0)  # Assuming no error if not provided

    for key, values in data.items():
        mean_array = np.array(values['mean'])
        err_array = np.array(values['err'])
        print(f"{key}_{name}_mean = np.array({mean_array.tolist()})")
        print(f"{key}_{name}_err = np.array({err_array.tolist()})")
        print("")
```

```python

# Path containing the output file

path = './results/sample_DOPED_BOW_Nup21_2D_8x8_SSH_U0_a08_o1.o13758495'

output_name = 'test_U0_a08_o1_Nup21'


```

```python
process_data(path, output_name)
```
